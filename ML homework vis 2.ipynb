```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('uber-raw-data-sep14.csv')
print(df.shape)
df.head()
```

```python
df['Lat'] = df['Lat'].apply(lambda x: np.floor(x*1000)/1000)
```

```python
df['Lon'] = df['Lon'].apply(lambda x: np.floor(x*1000)/1000)
```

```python
df_sample = df.sample(frac=0.3)
```

```python
plt.clf()
fig = sns.jointplot(x = df_sample['Lon'], y = df_sample['Lat'], kind = 'hex', size = 10)
plt.subplots_adjust(top=0.9)
fig.fig.suptitle('Uber sample location jointplot')
plt.show()
```

```python
fig.savefig('Uber Sample Location.png')
```

```python
df_zoom = df[(df['Lat']>40.6) & (df['Lat']<40.9) & (df['Lon']> -74.1) & (df['Lon'] < -73.7)]
df_zoom_sample = df_zoom.sample(frac = 0.3)
```

```python
plt.clf()
fig1 = sns.jointplot(x = df_zoom_sample['Lon'], y = df_zoom_sample['Lat'], kind = 'hex', size = 20)
plt.subplots_adjust(top=0.9)
fig1.fig.suptitle('Uber sample location jointplot (zoomed)')
plt.show()
```

```python
fig1.savefig('Uber Sample Location Zoomed.png')
```
