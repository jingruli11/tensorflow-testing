### Q2: Which 5 stocks have been growing the fastest in purchase volume? 

- There are 2 ways of defining growth:
    
    1. average incremental percent change 
    2. slope of linear model against time

Before fitting any model, we need to clean the data.

To start with, the time in df_buy is string. We need to change it to datetime object.

```python
# transform string to datetime object
df_buy['created_at'] = pd.to_datetime(df_buy['created_at'])
```

Take Mastercard (id = 83) as an example to preview the purchase data

```python
# preview mastercard volume
df_buy[df_buy['card_id'] == 83].head(20)
```

The purchase data happens throughout seconds, minutes and days. We need to calculate daily total volume for further analysis.

```python
# transform datetime to date and calculate daily volume
df_buy['date'] = df_buy['created_at'].apply(lambda x: x.toordinal())
buy_daily = df_buy.groupby(['card_id','date'])['value'].agg(sum)
df_buy_daily = buy_daily.reset_index(drop = False)
df_buy_daily.rename(columns={'value':'daily_volume'}, inplace = True)
# transform ordinal date to datetime
df_buy_daily['date'] = df_buy_daily['date'].apply(lambda x: dt.datetime.fromordinal(x))
```

```python
# plot of mastercard volume through time
plt.clf()
plt.figure(figsize = (10,6))
plt.suptitle('Mastercard volume over time', fontsize = 10)
# distribution of lag with raw data
plt.plot(df_buy_daily[df_buy_daily['card_id'] == 83]['date'], df_buy_daily[df_buy_daily['card_id'] == 83]['daily_volume'])
plt.show()
```

```python
# function to calculate average incremental percent change
def increment(df, df_cards):
    cards = df['card_id'].unique()
    result = []
    for i in cards:
        df_temp = df[df['card_id'] == i].sort_values('date', ascending = True)
        df_temp['pct_change'] = df_temp['daily_volume'].pct_change()
        result.append(df_temp['pct_change'].mean())
    df_raw = pd.DataFrame({'id':cards, 'avg_pct_change':result})
    df_output = df_raw.merge(df_cards[['id','name']], how = 'left', on = 'id')
    df_output.sort_values('avg_pct_change', ascending = False, inplace = True)
    return df_output
```

The following are the top 5 growth stocks based on incremental percentage change.

```python
# calculate average incremental change and show the top 5 stocks
df_top_pct_change = increment(df_buy_daily, df_cards)
df_top_pct_change.head()
```

```python
# plot top 5 growth stocks
plt.clf()
plt.figure(figsize = (16,10))
plt.suptitle('Top 5 growth stocks over time -- incremental percentage change', fontsize = 15)
# On Cloud Nine over time
plt.subplot(3,2,1)
plt.title('On Cloud Nine')
plt.plot(df_buy_daily[df_buy_daily['card_id'] == 19]['date'], df_buy_daily[df_buy_daily['card_id'] == 19]['daily_volume'])
# General Electric over time
plt.subplot(3,2,2)
plt.title('General Electric')
plt.plot(df_buy_daily[df_buy_daily['card_id'] == 50]['date'], df_buy_daily[df_buy_daily['card_id'] == 50]['daily_volume'])
# JD.com over time
plt.subplot(3,2,3)
plt.title('JD.com')
plt.plot(df_buy_daily[df_buy_daily['card_id'] == 84]['date'], df_buy_daily[df_buy_daily['card_id'] == 84]['daily_volume'])
# General Dynamics over time
plt.subplot(3,2,4)
plt.title('General Dynamics')
plt.plot(df_buy_daily[df_buy_daily['card_id'] == 96]['date'], df_buy_daily[df_buy_daily['card_id'] == 96]['daily_volume'])
# Bristol-Myers Squibb over time
plt.subplot(3,2,5)
plt.title('Bristol-Myers Squibb')
plt.plot(df_buy_daily[df_buy_daily['card_id'] == 113]['date'], df_buy_daily[df_buy_daily['card_id'] == 113]['daily_volume'])
plt.show()
```

```python
# function to calculate slope of linear model
def linear(df, df_cards):
    cards = df['card_id'].unique()
    result = []
    for i in cards:
        df_temp = df[df['card_id'] == i].sort_values('date', ascending = True)
        reg = linear_model.LinearRegression()
        reg.fit(df_temp[['date']], df_temp[['daily_volume']])
        result.append(reg.coef_[0][0])
    df_raw = pd.DataFrame({'id':cards, 'slope':result})
    df_output = df_raw.merge(df_cards[['id','name']], how = 'left', on = 'id')
    df_output.sort_values('slope', ascending = False, inplace = True)
    return df_output
```

The following are the top 5 growth stocks based on slopes over time

```python
# calculate slopes of each stock and show the result
df_top_slope = linear(df_buy_daily, df_cards)
df_top_slope.head()
```

```python
# plot top 5 growth stocks
plt.clf()
plt.figure(figsize = (16,12))
plt.suptitle('Top 5 growth stocks over time -- linear regression slope', fontsize = 15)
# Netflix over time
plt.subplot(3,2,1)
plt.title('Netflix')
plt.plot(df_buy_daily[df_buy_daily['card_id'] == 53]['date'], df_buy_daily[df_buy_daily['card_id'] == 53]['daily_volume'])
# Amazon over time
plt.subplot(3,2,2)
plt.title('Amazon')
plt.plot(df_buy_daily[df_buy_daily['card_id'] == 46]['date'], df_buy_daily[df_buy_daily['card_id'] == 46]['daily_volume'])
# Long-Term Mix over time
plt.subplot(3,2,3)
plt.title('Long-Term Mix')
plt.plot(df_buy_daily[df_buy_daily['card_id'] == 44]['date'], df_buy_daily[df_buy_daily['card_id'] == 44]['daily_volume'])
# Internet Titans over time
plt.subplot(3,2,4)
plt.title('Internet Titans')
plt.plot(df_buy_daily[df_buy_daily['card_id'] == 7]['date'], df_buy_daily[df_buy_daily['card_id'] == 7]['daily_volume'])
# Defending America over time
plt.subplot(3,2,5)
plt.title('Defending America')
plt.plot(df_buy_daily[df_buy_daily['card_id'] == 11]['date'], df_buy_daily[df_buy_daily['card_id'] == 11]['daily_volume'])
plt.show()
```

#### Data Cleaning

```python
# dataframe of user total volume per card
buy_users = df_buy.groupby(['user_id', 'card_id'])['value'].agg(sum)
df_buy_users = buy_users.reset_index(drop = False)
df_buy_users.rename(columns={'value':'volume'}, inplace = True)
# merge df_user with purchase data
df_temp1 = df_user.merge(df_buy_users, how = 'left', on = 'user_id')
```

#### Feature generation and validation

```python
# merge df_temp1 with cards data
df_cards.rename(columns={'risk_level':'card_risk_level', 'id':'card_id'}, inplace = True)
df_temp2 = df_temp1.merge(df_cards[['card_id', 'last_price', 'dividend_rate_percent', 'expense_ratio_percent',
                                    'one_day_change_percent', 'year_to_date_change_percent', 'card_risk_level',
                                    'start_of_year_price', 'start_of_year_price_year']], how = 'left', on = 'card_id')
```

```python
# merge df_temp2 with bookmark data
# create a new column "bookmark"
df_bookmark['bookmark'] = 1
df_temp3 = df_temp2.merge(df_bookmark[['user_id', 'card_id', 'bookmark']], how = 'left', on = ['user_id', 'card_id'])
# fill empty values with 0
df_temp3.fillna(0, inplace = True)
```

```python
# drop all keys
df_temp3.drop(columns=['user_id', 'uuid', 'card_id'], inplace = True)
# convert 'employment_status' to dummy values
for elem in df_temp3['employment_status'].unique():
    df_temp3[str(elem)] = df_temp3['employment_status'].apply(lambda x: 1 if x == elem else 0)
# drop 'employment_status' and 'RETIRED' to avoid multicolinearality
df_temp3.drop(columns=['employment_status','RETIRED'], inplace = True)
```

```python
# store Y labels
# the reason for using quartiles is to make each label volume relatively even
quantile1 = df_temp3['volume'].quantile(0.25)
quantile2 = df_temp3['volume'].quantile(0.5)
quantile3 = df_temp3['volume'].quantile(0.75)
def compare(x, q1, q2, q3):
    if x < q1:
        return 0
    elif x < q2:
        return 1
    elif x< q3:
        return 2
    else:
        return 3
    
Y = df_temp3['volume'].apply(lambda x: compare(x,quantile1, quantile2, quantile3))
```

```python
Y.value_counts()
```

```python
df_temp3.columns
```

```python
# test correlation among features
features_raw = df_temp3[['federal_tax_bracket_percent', 'liquidity_needs',
       'total_net_worth_range', 'age', 'risk_level', 'last_price',
       'dividend_rate_percent', 'expense_ratio_percent',
       'one_day_change_percent', 'year_to_date_change_percent',
       'card_risk_level', 'start_of_year_price', 'start_of_year_price_year',
       'bookmark', 'EMPLOYED', 'STUDENT', 'UNEMPLOYED']]
# print correlation matrix
features_raw.corr()
```

```python
# plot correlation matrix
corr = features_raw.corr()
plt.clf()
fig, ax = plt.subplots(figsize=(8, 8))
ax.matshow(corr)
plt.xticks(range(len(corr.columns)), corr.columns);
plt.yticks(range(len(corr.columns)), corr.columns);
plt.show()
```

Based on correlation analysis, the only problem within features is that 'start_of_year_price' and 'last_price' are highly correlated. Thus we drop 'start_of_year_price' to avoid multicollinearity.

```python
# drop 'start_of_year_price'
features_raw.drop(columns=['start_of_year_price'], inplace = True)
```

```python
features_raw.info()
```

```python
# normalize all numerical features
features = normalize(features_raw, norm='l2')
```

```python
# split training and testing data
X_train, X_test, y_train, y_test = train_test_split(features, Y, test_size=0.3, random_state=0)
```

```python
# print training and testing data size
print('Training data size')
print(X_train.shape, y_train.shape)
print()
print('Testing data size')
print(X_test.shape, y_test.shape)
```

#### PCA Analysis

```python
# Create a Randomized PCA model that takes two components
randomized_pca = PCA(n_components=2, svd_solver='randomized')

# Fit and transform the data to the model
rpca = randomized_pca.fit_transform(features)

# Inspect the shape
rpca.shape
```

```python
# print Y label size 
Y.count()
```

```python
# show PCA scatter plot
plt.clf()
colors = ['black', 'orange', 'red', 'green']
for i in range(4):
    x = rpca[:, 0][Y == i]
    y = rpca[:, 1][Y == i]
    plt.scatter(x, y, c=colors[i])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title("PCA Scatter Plot")
plt.show()
```

From the PCA plot, it is evident that the users who make little purchase (label = 0) are seperated from the others.

Meanwhile, the users making small to huge purchases seem to have similar characteristics based on existed data. That explains why they show up in the same clusters in PCA plot.

#### Logistic Regression Classification

```python
# try different parameters and tune Logistic Regression

start = timeit.default_timer()

param_logit = [0.1, 1, 10, 50]
train_scores_logit = []
test_scores_logit = []
logit_model = []

# Due to limited computing power, the parameters will be tested in sub-for-loops
for i in param_logit:
    clf_temp = LogisticRegression(penalty = 'l2', C = i, solver ='newton-cg')
    clf_temp.fit(X_train, y_train)
    temp_train_score = clf_temp.score(X_train, y_train)
    temp_test_score = clf_temp.score(X_test, y_test)
    train_scores_logit.append(temp_train_score)
    test_scores_logit.append(temp_test_score)
    logit_model.append(clf_temp.coef_)

stop = timeit.default_timer()
clf_tune_logit_time = stop - start
print('Running time: %.3f' % (stop - start))
```

```python
# create a dataframe to display comparison
d_tune_logit = {'C': param_logit,
            'Training Accuracy': train_scores_logit,
            'Testing Accuracy': test_scores_logit}
df_tune_logit = pd.DataFrame(data = d_tune_logit)
df_tune_logit.set_index('C', inplace = True)
```

```python
# display the result
plt.clf()

fig_logit = sns.heatmap(df_tune_logit, annot=True, fmt='f')
plt.subplots_adjust(top=0.9)
fig_logit.set_title('Logistic Regression Accuracy Heatmap')

plt.show()
```

```python
# best model's coefficients
logit_model[0][3]
```

```python
# The most important features:
print('The most important features from Logistic Regression:')
print()
print(features_raw.columns[3])
print(features_raw.columns[5])
print(features_raw.columns[9])
print(features_raw.columns[11])
```

#### Linear SVM

```python
# try different parameters and tune Linear SVM

start = timeit.default_timer()

param_lsvm = [0.01, 0.1, 1, 5]
train_scores_lsvm = []
test_scores_lsvm = []
lsvm_model = []
for i in param_lsvm:
    clf_temp = LinearSVC(penalty = 'l2', C = i, multi_class = 'ovr')
    clf_temp.fit(X_train, y_train)
    temp_train_score = clf_temp.score(X_train, y_train)
    temp_test_score = clf_temp.score(X_test, y_test)
    train_scores_lsvm.append(temp_train_score)
    test_scores_lsvm.append(temp_test_score)
    lsvm_model.append(clf_temp.coef_)

stop = timeit.default_timer()
clf_tune_lsvm_time = stop - start
print('Running time: %.3f' % (stop - start))
```

```python
# create a dataframe to display comparison
d_tune_lsvm = {'C': param_lsvm,
            'Training Accuracy': train_scores_lsvm,
            'Testing Accuracy': test_scores_lsvm}
df_tune_lsvm = pd.DataFrame(data = d_tune_lsvm)
df_tune_lsvm.set_index('C', inplace = True)
```

```python
# display the result
plt.clf()

fig_lsvm = sns.heatmap(df_tune_lsvm, annot=True, fmt='f')
plt.subplots_adjust(top=0.9)
fig_lsvm.set_title('Linear SVM Accuracy Heatmap')

plt.show()
```

```python
# best model's coefficients
lsvm_model[0][3]
```

```python
# The most important features:
print('The most important features from Linear SVM:')
print()
print(features_raw.columns[9])
print(features_raw.columns[11])
```

#### Random Forest Classification

```python
# try different parameters and tune random forest

start = timeit.default_timer()

param_rdmfr = [10, 25, 50, 100]
train_scores_rdmfr = []
test_scores_rdmfr = []
rdmfr_model = []
for i in param_rdmfr:
    clf_temp = RandomForestClassifier(n_estimators = i, criterion = 'entropy', max_depth = 5)
    clf_temp.fit(X_train, y_train)
    temp_train_score = clf_temp.score(X_train, y_train)
    temp_test_score = clf_temp.score(X_test, y_test)
    train_scores_rdmfr.append(temp_train_score)
    test_scores_rdmfr.append(temp_test_score)
    rdmfr_model.append(clf_temp.feature_importances_)

stop = timeit.default_timer()
clf_tune_rdmfr_time = stop - start
print('Running time: %.3f' % (stop - start))
```

```python
# create a dataframe to display comparison
d_tune_rdmfr = {'C': param_rdmfr,
            'Training Accuracy': train_scores_rdmfr,
            'Testing Accuracy': test_scores_rdmfr}
df_tune_rdmfr = pd.DataFrame(data = d_tune_rdmfr)
df_tune_rdmfr.set_index('C', inplace = True)
```

```python
# display the result
plt.clf()

fig_rdmfr = sns.heatmap(df_tune_rdmfr, annot=True, fmt='f')
plt.subplots_adjust(top=0.9)
fig_rdmfr.set_title('Random Forest Accuracy Heatmap')

plt.show()
```

```python
# best model's coefficients
[round(x,4) for x in rdmfr_model[3]]
```

#### Conclusion:

The key drivers for a user to make a purchase are:

- year_to_date_change_percent
- start_of_year_price_year
- federal_tax_bracket_percent
- total_net_worth
- risk_level
- last_price

I would recommend these 3 strategies:

1. Distinguish users based on their background information
    - From the PCA analysis we can see there are no distinct difference between users in purchasing volume. This means they all make small or big purchases at the same time. Since volume is not a good criteria to categorize users, we should use other information. For example, `federal_tax_bracket_percent`, `total_net_worth`, and `risk_level` are key drivers for users to purchase, and they are important features of the users themselves.

2. Put more weights in portfolio on certain high value stocks
    - From the previous question, we spotted that a lot of high growth stocks are high priced technology stocks, for example `Netflix` and `Amazon`. This immitates the actual trend in the market. The technology stocks have been rocketing in the past few years, and they have quite promising prospect. Recommending these stocks to clients would be easier to convince them to purchase.
    
3. Design a better system than bookmark to show what stocks users are interested in
    - One thing I thought would be key driver is whether the stock is bookmarked by the user. But this feature did not show up in any of those models. That means the current bookmark system shows little correlation with whether the users will make purchases. Thus a better system to keep track of user interests and make recommendations is necessary.

Other data I would want to improve the analysis:

- clarification on existing data:
    - volume means quantity or monetary value
    - what discover table means
- more user data to cluster user group, for example user education, marital status, number of children, debt amount, etc
- more detailed purchase data, for instance whether those purchases are individual or collective in a portfolio
